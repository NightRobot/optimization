```python
import random,time
import numpy as np
import pandas as pd
from itertools import combinations
from pprint import pprint
```

```python
def readData (file):
    xlsx = pd.ExcelFile(file)
    workload_sheet = []
    for sheet in xlsx.sheet_names:
        workload_sheet.append(xlsx.parse(sheet))
    return workload_sheet
```

```python
workload = readData("../data/workload_aug2018.xlsx")
print(workload)
```

```python
print(workload[0].keys())
keys = workload[0].keys().values.tolist()
print(keys)
print(keys[2:])
```
